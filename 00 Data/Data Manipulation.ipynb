```python
import pandas as pd
import os
```

```python
df_data = pd.DataFrame()

input_files = os.listdir("Data")

count = 1

for file in input_files:

    if file.endswith('.csv'):
        country = file.replace('.csv','')
        df_temp = pd.read_csv('Data\{}'.format(file),encoding='latin')
        df_temp['country'] = country
        
        df_data = df_data.append(pd.DataFrame(df_temp))
        
df_data.to_csv('Angelist Startups in Africa.csv')
```
